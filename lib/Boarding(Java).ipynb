```java
public boolean checkAsle(int[][] asle) {
    for (int[] loc : asle)
        if (loc[0] > -1)
            return true;
    return false;
}

public int seatingTime(int cols, int row, int col, int[][] seated) {
    int half = cols/2;
    int timing = 1;
    int start = col < half ? col+1 : half;
    int end = col < half ? half : col;
    for(int i=start; i<end; i++)
        if (seated[row][i]>0)
            timing += 3;
    return timing;
}

public int boarding(int rows, int cols, int[][] passengers){
    
    int duration = 0;
    int entryPassenger = 0;
    int[][] asle = new int[rows][];
    int[][] seated = new int[rows][cols];
    
    for(int i=0; i<rows; i++)
        asle[i] = new int[]{-1, -1, -1};

    while (entryPassenger < passengers.length || checkAsle(asle)) {
        for (int i = rows-1; i>-1; i--) {
            int row = asle[i][0];
            int col = asle[i][1];
            int lug = asle[i][2];
            if (row>-1) {                                 // check if current asle location has a person
                if (row == i) {                           // check if this person arrived his/her row
                    if (lug > 0)                          // check if this person stowed his/her luggage
                        asle[i][2]--;
                    else {                                // ready to seat
                        if (seated[row][col] > 1)         // check if this person currently in seating...
                            seated[row][col]--;           // in seating...
                        else if (seated[row][col] == 1) { // check if this person already seated
                            asle[i][0] = -1;              // mark asle empty
                            asle[i][1] = -1;
                            asle[i][2] = -1;
                        } else {                          // determine how long this person need for seating
                            seated[row][col] = seatingTime(cols, row, col, seated);
                        }
                    }
                }else if (asle[i+1][0]<0) {               // if next asle position is empty?
                    asle[i+1][0] = row;                   // move to next position...
                    asle[i+1][1] = col;
                    asle[i+1][2] = lug;
                    asle[i][0] = -1;
                    asle[i][1] = -1;
                    asle[i][2] = -1;
                }
            }
        }
        if (entryPassenger < passengers.length && asle[0][0] < 0)
            asle[0] = passengers[entryPassenger++];
        duration++;
    }
    return duration;
}
```
