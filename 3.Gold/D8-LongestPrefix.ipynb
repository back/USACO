# Section 2.3 - Longest Prefix
A special DP problem which asks for True/False value at each level...

### Read Inputs

```python
p = 'A AB BA CA BBC'.split()
s = 'ABABACABAABC'

print(p)
print(s)
```

### Prepare our working list
- Use 1-index based array
- For each indexed position:
  - 1: means reachable
  - 0: means un-reachable

```python
mark = [0] * (len(s)+1)
mark[0] = 1
print(mark)
```

---
Increamental Calculate through the Working List
---

```python
for i in range(len(mark)):
    if mark[i]:
        for n in p:
            if s[i:].startswith(n):     # if ( s.substring(i).startsWith(n) ) {...}
                mark[i + len(n)] = 1
print(mark)
```

---
Find the Final Answer
---

```python
# answer = max(idx for idx,value in enumerate(mark) if value)
answer = len(mark) - 1
while mark[answer] == 0:
       answer -= 1
print(answer)
```
