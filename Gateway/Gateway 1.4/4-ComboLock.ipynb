# Section 1.4 - Combination Lock
---
### Read inputs

```python
fin=StringIO('''50
1 2 3
5 6 7''')

n  = int(fin.readline().strip())
p1 = list(map(int, fin.readline().split()))
p2 = list(map(int, fin.readline().split()))

print (n, p1, p2)
```

---
### Search for all possible combinations
- Using MOD to normalize the wrapping around numbers

```python
k=set()                                     # HashSet<String> k = new HashSet<String>();
for r1 in range(-2, +3):                    # for (int r1 = -2; r1 < 3; r1++)    // gives us 5 loop: -2, -1, 0, 1, 2
    for r2 in range(-2, +3):
        for r3 in range(-2, +3):            # It will 5x5x5=125 combinations here for all possible keys
            
            k.add(((p1[0] + r1)%n, (p1[1] + r2)%n, (p1[2] + r3)%n))      # Using "mod" to bring it back in range!   [0 ~ n-1]
            k.add(((p2[0] + r1)%n, (p2[1] + r2)%n, (p2[2] + r3)%n))      # Using "mod" to bring it back in range!
            
print(k)
print(len(k))                               # k.size();
```
