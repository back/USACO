# Container with Most Water

https://leetcode.com/problems/container-with-most-water/

---
### Read Inputs

```python
height=[1,8,6,2,5,4,8,3,7]
print(height)
```

```python
answer, i, j = 0, 0, len(height)-1

while i<j:
    w = j - i
    
    if height[i] < height[j]:
        h=height[i]
        i+=1
    else:
        h=height[j]
        j-=1
        
    v = w * h
    
    if v > answer:
        answer = v
        
    print(i, j, v, answer)
```
